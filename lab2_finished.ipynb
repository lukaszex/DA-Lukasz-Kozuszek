```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plotHist (df, bins):
    df = pd.melt(df)
    g = sns.FacetGrid(df, col = 'variable', col_wrap = 2, sharex = False)
    g.map(plt.hist, 'value', bins = bins)
    g2 = sns.FacetGrid(df, col = 'variable', col_wrap = 2, sharex = False)
    g2.map(plt.hist, 'value', bins = bins, cumulative = True)
```

```python
size = 10000
distributions = {'N(0, 1)': np.random.normal(0, 1, size),
                 'N(0, 5)': np.random.normal(0, 5, size),
                 'P(1)': np.random.poisson(1, size),
                 'P(10)': np.random.poisson(10,size),
                 'L(0, 0.1)': np.random.lognormal(0, 0.1, size),
                 'L(0, 1)': np.random.lognormal(0, 1, size),
                 'B(1, 5)': np.random.beta(1, 5, size),
                 'B(5, 1)': np.random.beta(5, 1, size)
                 }
df = pd.DataFrame(distributions)
df.head()
```

```python
sns.set(style = 'darkgrid')
plotHist(df, 50)
```

```python
import scipy as sp
distributions = {'N(0, 1)': sp.special.expit(np.random.normal(0, 1, size)),
                 'N(0, 5)': sp.special.expit(np.random.normal(0, 5, size)),
                 'P(1)': sp.special.expit(np.random.poisson(1, size)),
                 'P(10)': sp.special.expit(np.random.poisson(10, size)),
                 'L(0, 0.1)': sp.special.expit(np.random.lognormal(0, 0.1, size)),
                 'L(0, 1)': sp.special.expit(np.random.lognormal(0, 1, size)),
                 'B(1, 5)': sp.special.expit(np.random.beta(1, 5, size)),
                 'B(5, 1)': sp.special.expit(np.random.beta(5, 1, size))
                 }
df2 = pd.DataFrame(distributions)
df2.head()
```

```python
plotHist(df2, 20)
```
